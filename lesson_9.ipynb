```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
x_data, y_data = datasets.fetch_california_housing(return_X_y=True, as_frame=True)
```

```python
#x = x_data.loc[:, ['MedInc', 'HouseAge'] ]
x = x_data.loc[:, ['MedInc'] ]
x
```

```python
y_data
```

```python
#arskirti duomenys
X_train, X_test, y_train, y_test = train_test_split(x, y_data, test_size=0.2)#test_size nuo 0 iki 1, shuffle - sumaisi duomenis
print(X_train.shape)
print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model = linear_model.LinearRegression()
```

```python
model.predict(X_test)# kalida, neapmokytas
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
#reikia paskaiciuot ar veikia gerai, paskaiciuoti paklaidas
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```

```python
print(f'train MSE: {mean_squared_error(y_train, pred_train)} ')
print(f'test MSE: {mean_squared_error(y_test, pred_test)} ')
```

```python
print(f'train R^2: {model.score(X_train, y_train)}')
print(f'test R^2: {model.score(X_test, y_test)}')
```

```python
plt.scatter(X_train, y_train)
plt.scatter(X_train, pred_train, c='r')
plt.scatter(X_test, y_test, c = 'g')
#plt.scatter(X_train['MedInc'], X_train['HouseAge'])
```
