```python
# Pervadinu stulpelius
df = df.rename( columns={'Entity': 'Country'} )
df
```

```python
df['GDP/Capita'] # paziuriu stulpeli su BVP
```

```python
# noriu unikaliu salies pavadinimu, nepriklausomai nuo duomenu
unique_countries = df['Country'].unique()

print(unique_countries)

#df_1 = df.sort_values(by=['Country']).drop_duplicates(subset='Country') # unikalios salis
#df_1
```

```python
df_sum = df.groupby('Country').agg( {'GDP/Capita': 'max', 'Population': 'max', 'Daily tests': 'sum', 'Cases': 'sum', 'Deaths': 'sum'} )# susumuoju testu, susirgimu ir mirciu skaicius, sugrupuoju pagal sali
df_sum = df_sum.sort_values('GDP/Capita',ascending=False)# surusiuoju pagal BVP mazejimo tvarka
df_sum
```

```python
# pagalvojau kad nesamoningai siaip suskaiciuoti mirciu ir susirgimu skaiciu, nes dar skiriasi saliu populiacijos
df_sum['Percent Cases'] = (df_sum['Cases'] / df_sum['Population'])* 100 # bandau procentaliai suskaiciuoti susirgimu atv pagal populiacija
df_sum = df_sum.round(2) #suapvalinau iki 2 skaiciu po kablelio
df_sum
# gavosi idomus "Percent Cases", pvz. 1070.59
# Stebint Luxembourg pvz gaunasi labai didelis Percent Cases (1 zmogus susirgo 10 kartu per metus). 
# Tai arba duomenys klaidingi, arba salys norejo gauti finansines ar dar kokios naudos.
# Arba kazka netaip skaiciuoju/galvoju
```

```python
df_sum['Percent Deaths'] = (df_sum['Deaths'] / df_sum['Population'])* 100 #  procentaliai mirciu sk pagal populiacija
df_sum = df_sum.round(2) #suapvalinau iki 2 skaiciu po kablelio
df_sum
```

```python
df_sum=df_sum.drop('GDP/Capita', axis=1)# istriniau BVP stulpeli
df_sum
```

```python
def calc_correlation(df_sum):
    correlation = df_sum['Daily tests'].corr(df_sum['Cases'])
    return correlation
correlation = calc_correlation(df_sum)
correlation
```

```python
"""
Ar vidutinė temperatūra šalyje turi įtakos COVID-19 atvejų ir mirčių skaičiui? Teigiu, kad kuo aukštesnė vid. temperatūra, tuo mažiau susirgimų ir mirčių.
"""
```

```python

df_temp = df.groupby('Country').agg( {'Average temperature per year': 'max', 'Population': 'max', 'Daily tests': 'sum', 'Cases': 'sum', 'Deaths': 'sum'} )
df_temp = df_temp.sort_values('Average temperature per year',ascending=False)
df_temp['Percent Cases'] = (df_temp['Cases'] / df_temp['Population'])* 100 #  procentaliai mirciu sk pagal populiacija
df_temp = df_temp.round(2) #suapvalinau iki 2 skaiciu po kablelio
df_temp['Percent Deaths'] = (df_temp['Deaths'] / df_temp['Population'])* 100 #  procentaliai mirciu sk pagal populiacija
df_temp = df_temp.round(2) #suapvalinau iki 2 skaiciu po kablelio
df_temp
```

```python
df_temp.head(20)
```

```python
df_temp.tail(20)
```

```python
def calc_correlation1(df_temp):
    correlation = df_temp['Average temperature per year'].corr(df_temp['Cases'])
    return correlation
correlation1 = calc_correlation1(df_temp)
correlation1
```
